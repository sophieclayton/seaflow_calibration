```python
% matplotlib inline

import numpy as np
import pandas as pd
import seaflowpy as sfp

import matplotlib.pyplot as plt
```

```python
bead_data = sfp.EVT('../data/calsettings10_13_2015.evt', transform=False, columns=["fsc_small", "D1","D2"])

bead_data.df.info()
```

```python
# try to fit a polynomial to bead data
z = np.polyfit(bead_data.df['fsc_small'], bead_data.df['D1'], 3)
p = np.poly1d(z)
xp = np.linspace(0, 6e04, 100)
```

```python
fig1 = plt.figure(figsize=(20,20))

plt.scatter(bead_data.df['fsc_small'], bead_data.df['D1'], marker ='.')
plt.plot(xp, p(xp),'.-r')
plt.show()
```
